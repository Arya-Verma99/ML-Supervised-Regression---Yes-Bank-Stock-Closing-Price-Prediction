#### Chart - 3

````python
# Plotting graph Independent variable vs Dependent variable to check Multicollinearity.

# Create subplots: adjusting rows and columns depending on the number of variables
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Loop through each column except the last one (as specified by columns[:-1])
for i, col in enumerate(columns[:-1]):
    ax = axes[i]  # Select the correct subplot

    feature = df[col]
    label = df["Close"]
    correlation = feature.corr(label)

    # Scatter plot
    ax.scatter(x=feature, y=label)
    ax.set_ylabel("Closing price")
    ax.set_xlabel(col)
    ax.set_title(f'Closing price vs {col}, Correlation: {correlation:.2f}')

    # Add regression line
    z = np.polyfit(df[col], df['Close'], 1)  # Fit a 1st degree polynomial (linear fit)
    y_hat = np.poly1d(z)(df[col])            # Calculate the y-values for the regression line
    ax.plot(df[col], y_hat, "r--", lw=1)     # Plot the regression line

plt.tight_layout()
plt.show()
````

````python
# Getting the best parameter
lasso_regressor.best_params_
````

````python
# Getting the best score
lasso_regressor.best_score_
````
